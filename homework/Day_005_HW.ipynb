```python
#找出所有數值型column
quantitative = [f for f in app_train.columns if app_train.dtypes[f] != 'object']
quantitative
```

```python
#找出所有類別型column
qualitative = [f for f in app_train.columns if app_train.dtypes[f] == 'object']
qualitative
```

```python
print(app_train.columns)
#數值column資料的觀察
f = pd.melt(app_train, value_vars=quantitative)
print(f.head())
g = sns.FacetGrid(f, col="variable",col_wrap=5, sharex=False, sharey=False)
g = g.map(sns.distplot, "value")
```

```python
app_train[quantitative].mean(axis=0, skipna=True)
```

```python
app_train[quantitative].std(axis=0, skipna=True)
```

```python
app_train[quantitative].cov()
```

```python
#遺失值的觀察
missing = app_train.isnull().sum()
missing = missing[missing > 0]

missing.sort_values(inplace=True)
missing = missing/len(app_train.index)
missing.plot.bar()
```
